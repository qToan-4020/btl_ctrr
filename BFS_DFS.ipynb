# BFS(tm kim theo chiu rng)


```python
from collections import deque
# =========================
# Hm BFS
# =========================
def bfs(graph, start):
    # Tp nh du cc nh  thm
    visited = set()

    # Hng i dng cho BFS
    queue = deque()

    # nh du nh bt u
    visited.add(start)
    queue.append(start)

    print("Thu tu duyet BFS:")

    # Lp cho n khi queue rng
    while queue:
        # Ly nh  u queue
        current = queue.popleft()
        print(current, end=" ")

        # Duyt cc nh k
        for neighbor in graph[current]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    print()  # Xung dng cho p


# =========================
# Hm main
# =========================
if __name__ == "__main__":
    graph = {}

    # Nhp s nh v s cnh
    n = int(input("Nhap so dinh: "))
    m = int(input("Nhap so canh: "))

    # Khi to danh sch k rng cho mi nh
    print("Nhap ten cac dinh:")
    for _ in range(n):
        v = input().strip()
        graph[v] = []

    # Nhp cc cnh
    print("Nhap cac canh (u v):")
    for _ in range(m):
        u, v = input().split()

        #  th v hng
        graph[u].append(v)
        graph[v].append(u)

    # Nhp nh bt u BFS
    start = input("Nhap dinh bat dau BFS: ")

    # Gi BFS
    bfs(graph, start)



```

# DFS
